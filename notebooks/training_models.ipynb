```python
import os
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
import pickle
import json

df_brn = pd.read_csv("C:/Machine Learning/RuralAreaDiseasePredictor/datasets/brn.csv")

df_brn = df_brn.drop(columns=['id'])

df_brn = df_brn[sorted(df_brn.columns)]

encoders = {}
for col in df_brn.select_dtypes(include=['object']):
    encoder = LabelEncoder()
    df_brn[col] = encoder.fit_transform(df_brn[col])
    encoders[col] = encoder

df_brn.dropna(inplace=True)

rf = RandomForestClassifier(max_depth=7, n_estimators=50)

X = df_brn.loc[:, df_brn.columns != 'stroke']
y = df_brn['stroke']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

with open("C:/Machine Learning/RuralAreaDiseasePredictor/models/brn.pkl", "wb") as f:
    pickle.dump(rf , f)
with open("C:/Machine Learning/RuralAreaDiseasePredictor/encoders/brn_encoders.pkl","wb") as f:
    pickle.dump(encoders , f)

attributes = {}
attributes['brn'] = list(X_train.columns)
```

```python
import os
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
import pickle
import json

# Load the dataset
df_ast = pd.read_csv("C:/Machine Learning/RuralAreaDiseasePredictor/datasets/ast.csv")

# Sort the columns alphabetically
df_ast = df_ast[sorted(df_ast.columns)]

# Initialize a dictionary to store the encoders
encoders = {}

# Apply LabelEncoder to categorical columns
for col in df_ast.select_dtypes(include=['object']):
    encoder = LabelEncoder()
    df_ast[col] = encoder.fit_transform(df_ast[col])
    encoders[col] = encoder

# Drop missing values
df_ast.dropna(inplace=True)

# Initialize the RandomForestClassifier
rf = RandomForestClassifier(max_depth=7, n_estimators=50)

# Define X (features) and y (target variable)
# Correctly exclude the target columns from X
X = df_ast.drop(columns=['Severity_Moderate', 'Severity_None', 'Severity_Mild'])
y = df_ast['Severity_Moderate']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the model
rf.fit(X_train, y_train)

# Make predictions
y_pred = rf.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# Print classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Save the trained model to a pickle file
with open("C:/Machine Learning/RuralAreaDiseasePredictor/models/ast.pkl", "wb") as f:
    pickle.dump(rf, f)

# Save the encoders to a pickle file
with open("C:/Machine Learning/RuralAreaDiseasePredictor/encoders/ast_encoders.pkl", "wb") as f:
    pickle.dump(encoders, f)

# Save the feature names (columns of X_train) in the attributes dictionary
attributes['ast'] = list(X_train.columns)
print(attributes['ast'])
```
